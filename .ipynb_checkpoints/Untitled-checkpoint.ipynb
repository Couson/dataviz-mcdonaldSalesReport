```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import json
```

```python
monthly_sales = pd.read_csv('monthly_sales.csv')
monthly_sales.head()
```

```python
daily_sales = pd.read_csv('daily_sales.csv')
daily_sales.head()
```

```python
regions  = [r'NE', r'NW', r'SE', r'SW', r'(C$)']

items = [r'HM', r'CF', r'FF']

def filter_columns(df, to_filter):
    out = []
    for col in df.columns:
        if re.findall(to_filter, col):
            out.append(col)
            
    cal_col = df.loc[:, out].apply(np.sum, axis = 1)
    return cal_col

def classify_by(df, reference):
    
    out_df = pd.DataFrame()
    out_df.loc[:, df.columns[0]] = df.loc[:, df.columns[0]]

    for i in reference:
        cal_ser = filter_columns(df, i)
        out_df.loc[:, i] = cal_ser
    
#     print(df)
    out_df = out_df.rename(columns = {r'(C$)': 'C', 
                                  'Month, Year': 'date',
                                  'Unnamed: 0': 'date'})
    
    return out_df
```

```python
monthly_sales_byItem = classify_by(monthly_sales, items)
monthly_sales_byRegion = classify_by(monthly_sales, regions)

daily_sales_byItem = classify_by(daily_sales, items)
daily_sales_byRegion = classify_by(daily_sales, regions)
```
